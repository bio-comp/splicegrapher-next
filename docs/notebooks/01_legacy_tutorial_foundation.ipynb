# SGN Tutorial 01: Legacy Workflow Foundation

This notebook is the execution-safe foundation for the legacy-parity tutorial track.
It mirrors the original SpliceGrapher tutorial framing while using modernized SGN
configuration/data manifests.

References:

- Rogers et al. Genome Biology 2012 (SpliceGrapher)
- Rogers et al. BCB 2013 (SpliceGrapherXT)


```python
from pathlib import Path

try:
    import tomllib
except ModuleNotFoundError:  # pragma: no cover - Python < 3.11 fallback
    import tomli as tomllib


def _resolve_repo_root_and_manifest(start_dir: Path) -> tuple[Path, Path]:
    for base in [start_dir, *start_dir.parents]:
        candidate = base / "docs" / "tutorials" / "dataset_manifest.toml"
        if candidate.exists():
            return base, candidate
    raise FileNotFoundError(
        "Could not locate docs/tutorials/dataset_manifest.toml from cwd hierarchy."
    )


repo_root, manifest_path = _resolve_repo_root_and_manifest(Path.cwd().resolve())
with manifest_path.open("rb") as handle:
    manifest = tomllib.load(handle)

manifest["manifest"]
```

```python
species = manifest["species"]
for species_key, species_cfg in species.items():
    print(
        f"{species_key}: {species_cfg['name']} ({species_cfg['study_accession']}) "
        f"runs={len(species_cfg['run_accessions'])}"
    )
```

```python
import matplotlib

matplotlib.use("Agg")
import matplotlib.pyplot as plt

labels = []
run_counts = []

for species_key, species_cfg in species.items():
    labels.append(species_key)
    run_counts.append(len(species_cfg["run_accessions"]))

fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(labels, run_counts, color=["#2a9d8f", "#e76f51"])
ax.set_title("Canonical tutorial runs per species")
ax.set_ylabel("Run count")
for idx, count in enumerate(run_counts):
    ax.text(idx, count + 0.02, str(count), ha="center")
fig.tight_layout()
fig
```
